## Why Matplotlib?

* _Python_? Pros: Easy to use, lots of tools.
* _Matplotlib_?
 - Pros: 
   + Easy to start with, great for advance users.
   + Massive library: someone has already done it.
   + Concise: one-liners for basic plots.
 - Cons: 
   + Sometimes it's difficult to get perfect plots.
   + Some parts of the documentation are outdated.

Explore a dataset with some cell magic:

```python
%matplotlib notebook
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                style="island",
                hue='species',
                data=penguins)
```

```python
%matplotlib inline
```

```python
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
```

```python
sns.jointplot(data=penguins,
              x="bill_length_mm",
              y="bill_depth_mm",
              hue="species")
```
