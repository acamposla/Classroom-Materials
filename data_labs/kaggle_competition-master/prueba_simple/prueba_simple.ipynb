```python
data_train_bien = pd.get_dummies(train)
```

```python
data_train_bien
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
columnas_x = [a for a in data_train_bien.columns if a!= "price"]
columnas_x
```

```python
X = data_train_bien[columnas_x]
y = data_train_bien["price"]
```

```python
rf_reg = RandomForestRegressor(n_estimators=27, max_depth=3, min_samples_leaf=3, random_state=111)
```

```python
rf_reg.fit(X,y)
```

```python
predict = pd.read_csv("data/predict.csv")
```

```python
predict_bien = pd.get_dummies(predict)
```

```python
predict_bien.head()
```

```python
predict["price"] = rf_reg.predict(predict_bien)
```

```python
predict.head()
```

```python
para_exportar = predict[["id", "price"]]
para_exportar
```

```python
para_exportar.to_csv("prueba", header=True, index=False)
```
